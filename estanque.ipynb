```python
from pylab import imshow, show, xlabel, ylabel,plot,xlim,ylim
from numpy import loadtxt,linspace,sin,sqrt
import numpy as np
x=linspace(0,500, 10)
y=linspace(0,500,10)
#xlim(0,500)
#ylim(0,500)
c1=1 #amplitud
c2=1
b=5 #longitud de onda
k=(2*3.1416)/b
x1=50 #psicion inicial
y1=25 #posicion inicial
x2=50
y2=75
dimx=100
nx=500
dx=dimx/nx
salida=open("ondas_estanque.dat","w")
for i in range(0,501):
    for j in range(0,501):
        r1=sqrt((x1-i*dx)**2 +(y1-j*dx)**2)
        r2=sqrt((x2-i*dx)**2 +(y2-j*dx)**2)
        C=c1*sin(k*r1)+c2*sin(k*r2)
        salida.write("% 0.2f "% (C))
    salida.write("\n")
salida.close()
data = loadtxt("ondas_estanque.dat", float)
print(data.shape)
imshow(data, origin="lower", extent=[0, 100, 0, 100])
show()
```
